```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv("/content/detailed_crop_yield_predictions.csv")
file.head()
```

```python
file.info()
```

```python
file.describe()
```

```python
file.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
categorical_columns = ['Crop', 'Region', 'Fertilizer Used','Soil Quality']

for col in categorical_columns:
    file[col] = label_encoder.fit_transform(file[col])
```

```python
file.drop(columns=['Prediction Date'],inplace=True)
```

```python
file.head()
```

```python
from sklearn.model_selection import train_test_split
x = file.drop(columns=["Predicted Yield (tons/ha)"])
y = file["Predicted Yield (tons/ha)"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(ytest, ypred)
mae = mean_absolute_error(ytest, ypred)
r2 = r2_score(ytest, ypred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R2 Score:", r2)
```

```python
data = np.array([[1,3353, 28.66,706.50, 1, 3, 11]])
scaled_data = scaler.transform(data)
predictions = model.predict(scaled_data)

print(predictions)
```
